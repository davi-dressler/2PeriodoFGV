```python
df = pd.read_json("aula_pinho.json") #Leitura de arquivo JSON
df
```

```python
df["month"] = pd.to_datetime(df["month"])
df.dtypes
```

```python
df = df.set_index("month").sort_index()

ax = df.plot(figsize= (8, 4), title= "Evolução mensa de visualizações")
ax.set_xlabel("Mês")
ax.set_ylabel("Número de visualizações")
plt.show()
```

```python
totais = df.sum(axis= 0).sort_values(ascending= False)
ax = totais.plot(kind= "bar", rot= 45, title= "Total de visualizações no período por série")
ax.set_xlabel("Série")
ax.set_ylabel("Total de visualizações")
plt.show()
```

```python
medias = df.mean(axis= 0).sort_values(ascending= False)
ax = medias.plot(kind= "barh", title= "Média mensal de visualizações por série")
ax.set_xlabel("Série")
ax.set_ylabel("Média")
plt.show()
```

```python
top3 = totais.head(3)
ax = top3.plot(kind= "pie", ylabel= "", autopct= "%1.1f", title= "Top 3 séries mais assistidas no período")
plt.show()
```

```python
ax = df["Dexter"].plot(kind= "hist", bins = 5, title= "Histograma de visualizações de Dexter")
ax.set_xlabel("Visualizações")
plt.show()
```

```python
ax = df.plot(kind= "box",  title= "Boxplot das visualizações")
plt.show()
```

```python
ax = df.plot(kind= "scatter", x= "Breaking Bad", y= "Dexter", title= "Dispersão entre Dexter e Breaking Bad")
plt.show()
```
